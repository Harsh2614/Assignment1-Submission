**Import the Data**

```
da = pd.read_csv("brand_details.csv")
pd.set_option("display.max_rows",da.shape[0]+1)
```

**Problem1 - Which industry has the maximum brands whose Brand_Revenue > 80 million**
[ ]


```
da[da['Brand_Revenue']>80].groupby(['Industry'])['Brand'].count() 

```
