```
import pandas as pd
import numpy as np
df_brand = pd.read_csv('brand_details.csv',sep='\t')
pd.set_option('display.max_rows',df_brand.shape[0]+1)
df_brand.head()
```

```
#Problem 1 - Which industry has the maximum brands whose Brand_Revenue> 80 Million (Brand revenue is in million
df = df_brand[df_brand['Brand_Revenue'] > 80.0 ]
df_new = df.groupby(['Industry'],as_index=False)['Brand'].count()
df_new.loc[df_new['Brand'] == df_new["Brand"].max()]
```

```
#Problem 2 - List the companiesd whose Company_Advertising budget is < 3 but Brand_Value is > 20
df_brand[(df_brand['Company_Advertising'] < 3.0) & (df_brand['Brand_Value'] > 20.0)]
```

```
#Problem 3 - Find out the top 5 Technology Brand that showed maximum positive jump in their position (Use Change_LastYr)
df_brand[(df_brand['Industry'] == 'Technology') & (df_brand['Change_LastYr'] > 0.0)].nlargest(5,'Change_LastYr')
```

```
#Problem 4 - Find out the Average Brand_Revenue by Industry
df_avg = df_brand.groupby(['Industry'],as_index=False)['Brand_Revenue'].mean()
df_avg
```
